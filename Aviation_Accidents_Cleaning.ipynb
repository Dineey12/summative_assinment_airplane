### STEP4:Cleaning and Constructing Key Measurables

Main focus:
1. Construct metric for fatal/serious injuries.
2. Aircraft.Damage

#####  Construct metric for fatal/serious injuries.

```python
# Impute missing injury and passenger counts with 0
injury_cols = ['total.fatal.injuries', 'total.serious.injuries', 'total.minor.injuries', 'total.uninjured']
df[injury_cols] = df[injury_cols].fillna(0)
```

```python
# Estimate total passengers
df["total_passengers"] = df[injury_cols].sum(axis=1)
```

```python
# Construct metric for fatal/serious injuries fraction
df["fatal_serious_fraction"] = np.where(
    df["total_passengers"] > 0,
    (df["total.fatal.injuries"] + df["total.serious.injuries"]) / df["total_passengers"],
    0
)
```

```python
# Clean aircraft.damage column and fill it with unknown
df['aircraft.damage'] = df['aircraft.damage'].fillna("Unknown")
```

```python
#create a destroyed derived column (1 if destroyed, 0 otherwise)
df["is_destroyed"] = np.where(df['aircraft.damage'] == "Destroyed", 1, 0)
```

```python
#checking the aircraft.damage column with is_destroyed column
df[['aircraft.damage','is_destroyed']]
```

### STEP7:Cleaning Other Columns
There are other columns containing data that might be related to the outcome of an accident. We list a few here:

1. Engine.Type
2. Weather.Condition
3. Number.of.Engines
4. Purpose.of.flight
5. Broad.phase.of.flight
6. Inspect and identify potential cleaning tasks in each of the columns. Execute those cleaning tasks.



```python
#clean the categorical text columns
text_cols = ['engine.type', 'weather.condition', 'purpose.of.flight', 'broad.phase.of.flight']

for col in text_cols:
    #convert to Title Case (e.g., 'vMC' becomes 'Vmc') and remove spaces
    df[col] = df[col].str.title().str.strip()
    
    #replace 'Unk' or empty strings with 'Unknown'
    df[col] = df[col].replace(['Unk', '', 'None'], 'Unknown')
```

```python
#fix the 'Number.of.Engines' column
#some rows might have 0.0 or 10.0 (unlikely for standard insurance models)
#we force it to be numeric and cap it at a reasonable range
df['number.of.engines'] = pd.to_numeric(df['number.of.engines'], errors='coerce')
#replace 0 engines with NaN because an airplane must have at least 1 engine to be in this study
df.loc[df['number.of.engines'] < 1, 'number.of.engines'] = np.nan
```

```python
#inspection to see the most common values now
for col in text_cols:
    print(f"Top 5 values in {col}")
    print(df[col].value_counts().head(5))
    print("\n")
```
